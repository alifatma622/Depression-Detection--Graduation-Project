```python

# here is the function for reading the relative paths of all the images so that we can compare the paths with each other 
# to know whether they are similar or not 

def absoluteFilePaths( directory  ):
    
    for dirpath,_,filenames in os.walk(directory):                   # walk to go through all the folders 
        for f in filenames:
            yield os.path.relpath(os.path.join(dirpath, f))  # relpath to read the relative paths of all images 
            

```

```python

# here is we call the absoluteFilePaths function and give it the paths of our two folders training and testing 
# to read all the paths. 

all_training_paths =   list ( absoluteFilePaths( training_path )   )  # convert it list because the function returns generator

all_testing_paths =    list( absoluteFilePaths( testing_path )   )  # convert it list because the function returns generator
 

    
```

# Plotting Confusion Matrix 

```python

def drawing_conf_matrix(  y_actual , y_predict ):
    
    print( "Classification Report" )
    print ( classification_report(  y_actual , y_predict  )  )  # to represent accuacy, precision, recall and f1 score

    conf_mat = confusion_matrix( y_actual,   y_predict )   # to represent confusion matrix ( TP , FP , FN , FN )
    
    print( "Confusion Matrix")
    print( conf_mat )
    sns.heatmap( conf_mat , annot= True  , fmt = "" )
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# VGG-19 Trained Model

```python


from tensorflow.keras.applications import VGG19
 
vgg_model = VGG19(weights='imagenet',
                  include_top=False,
                  input_shape=(64, 64, 3))


```

```python

print(vgg_model.summary())
```

```python
vgg_model.layers[-1]         # showing the last layer  // MaxPooling Layer
```

```python


for layer in vgg_model.layers:
    layer.trainable = False

# We will have to use the functional API 

# last layers output
x = vgg_model.layers[ 6 ].output      #     to use the first 2 blocks of VGG-16

x= Conv2D( 128 , ( 3 , 3), padding="same", activation="relu")(x) # Adding a convolution layer with 32 filter of size 3*3
x=BatchNormalization()(x)
x=Dropout( 0.3 )(x)

x=MaxPooling2D(pool_size=(2, 2))(x)   #  Maxpooling layer to reduce the size 



x= Conv2D( 64 , ( 3 , 3), padding="same", activation="relu")(x) # Adding a convolution layer with 32 filter of size 3*3
x=Dropout( 0.3 )(x)

x=MaxPooling2D(pool_size=(2, 2))(x)   #  Maxpooling layer to reduce the size 

x= Conv2D(32, (3, 3), padding="same", activation="relu")(x)   # Adding a convolution layer with 32 filter of size 3*3
x=Dropout( 0.3 )(x)

x=MaxPooling2D(pool_size=(2, 2))(x)  #  Maxpooling layer to reduce the size 

x = Flatten()(x)   # to convert into 1D layer 

x=Dropout( 0.3 )(x)
x = Dense( 256, activation='relu')(x)    # adding fully connected layer with activation function = "relu"

x = Dense(2, activation='softmax')(x)    # adding fully connected layer with activation function = "softmax"


```

```python
 
from tensorflow.keras.models import Model
vgg_model_transfer = Model( inputs=vgg_model.input , outputs=x )
```

```python
print(vgg_model_transfer.summary())
```

```python
nClasses = 2
# Use Keras' handy utils
y_train_k = tensorflow.keras.utils.to_categorical( y_train, num_classes=nClasses)
y_valid_k = tensorflow.keras.utils.to_categorical( y_valid , num_classes=nClasses)
 
```
